```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### (1) 데이터 가져오기

```python
from sklearn.datasets import load_diabetes
diabetes=load_diabetes()
df_x = diabetes.data
df_y = diabetes.target
```

### (2) 모델에 입력할 데이터 X 준비하기

```python
x = df_x 
```

### (3) 모델에 예측할 데이터 y 준비하기

```python
y = df_y
```

### (4) train 데이터와 test 데이터로 분리하기

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, 
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=7) 
```

### (5) 모델 준비하기

```python
diabetes.feature_names
```

```python
# w, b 준비
W = np.random.rand(10)*10
b = np.random.rand()
W,b 
```

Model에서 사용할 W(Weight) 가중치와 b(bias) 편향 혹은 민감도를 랜덤 값으로 설정하였다.    
초기에 W와 b값은 아무 값이나 설정되어도 상관이 없다. 왜냐하면 처음에 엉뚱한 값을 주어도    
결국 경사하강법에 의해 적절한 W와 b찾아가기 때문이다.    
지금은 rand(10) 랜덤값에 *10을 해서 좀 더 극적인 Loss함수의 하강을 유도했지만, 10을 곱하지 않아도 상관없다.    

```python
#모델 함수 구현하기 
def model(X, W, b):
    predictions = 0
    for i in range(10):
        predictions += X[:, i] * W[i]
    predictions += b
    return predictions
```

model의 목적은 X에서의 미분값을 prediction = WX + b를 계산하여 prediction을 구하는 것이다.    
정확도가 높은 prediction을 산출하기 위해서는 경사하강법상에서 Loss가 가장 낮은 지점을 찾아가야 한다.    
그러기 위해서 W와 b값을 어떻게 찾는냐가 중요한 문제이다. 

### (6) 손실함수 loss 정의하기

```python
def MSE(a, b):
    mse = ((a - b) ** 2).mean()  # 두 값의 차이의 제곱의 평균
    return mse
```

MSE ( Mean Squared Error) 평균제곱오차 
<img src="https://user-images.githubusercontent.com/95423215/226958840-e80b70e1-0566-4035-a268-3584bd2fa6f7.png" width="20%" height="20%"/>
* 실제값과 예측값의 차이를 제곱해 구한 값의 평균값
* 실제값과 예측값 차이의 면적의 평균과 같다.
* 이상값이 존재하면 평균과의 차이가 커서 수치가 클 수 있다. 
* 0에 가까울 수록 실제값에 가까워 정확도가 높다고 할 수 있다. 

```python
def loss(X, W, b, y):
    predictions = model(X, W, b)   
    L = MSE(predictions, y)
    return L
```

Loss함수: 모델이 얼마나 좋은지를 평가하는 지표

* model에서 산출한 preditions를 이용해 MSE로 오차의 정도를 계산
* MSE, RMSE, Cross-Entropy, Hinge 등

### (7) 기울기를 구하는 gradient 함수 구현하기

```python
def gradient(X, W, b, y):
    # N은 데이터 포인트의 개수
    N = len(y)
    
    # y_pred 준비
    y_pred = model(X, W, b)
    
    # 공식에 맞게 gradient 계산
    dW = 1/N * 2 * X.T.dot(y_pred - y)
        
    # b의 gradient 계산
    db = 2 * (y_pred - y).mean()
    return dW, db
```

경사하강법(Gradient Descent)에 의해 Loss를 최소로 하는 지점을 찾기 위해 X에서 기울기 계산 (편미분) 
산출된 dW db는 Learning Rate와 계산되어 진다. 

결국 Gradiend의 값은 기울기의 방향을 결정하고, Learning_rate는 다음지점을 찾아가는 업데이트 크기를 조정한다. (1)


### (8) 하이퍼 파라미터인 학습률 설정하기

```python
LEARNING_RATE = 0.8
```

### (9) 모델 학습하기

```python
losses = []

for i in range(1, 10001):
    dW, db = gradient(X_train, W, b, y_train)
    W -= LEARNING_RATE * dW
    b -= LEARNING_RATE * db
    L = loss(X_train, W, b, y_train)
    losses.append(L)
    if i % 100 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

Gradient에서 산출된 dw와 db에 Learning_Rate를 곱해서 W와 b를 산출한다.    
이것은 for반복을 할 때마다 기울기가 낮은쪽을 향해 계속 수정되며 Loss가 가장 낮은 지점으로 찾아간다. 

```python
plt.plot(losses)
plt.show()
```

Loss가 초기에 높다가 2866정도에서 수평을 유지하며 더 이상 하강하지 않음을 볼 수 있다. 


### (10) test 데이터에 대한 성능 확인하기

```python
prediction = model(X_test, W, b)
mse = loss(X_test, W, b, y_test)
mse
```

### (11) 정답 데이터와 예측한 데이터 시각화하기

```python
plt.scatter(X_test[:, 0], y_test)
plt.scatter(X_test[:, 0], prediction)
plt.show()
```

## 회고

* 이번 [E-02]Regression을 통해서 머신러닝에서의 Model, Model을 가지고 학습한다는 것의 의미, Loss함수의 역할에 대해 핵심을 간명하게 확인 할 수 있었다. 
* Model: 가중치(W)와 편향(b)를 가지고 예측값 Prediction을 산출하는 함수
* Model을 학습한다는 의미: Prediction의 확률을 높이기 위해서 W와 b를 찾는 것
* Loss함수의 역할: Prediction이 실제값과 얼마나 차이가 있는지 알려주는 함수, 이것을 통해 학습을 계속해야 하는지, Learning Rate를 얼마나 조정할지에 대한 판단할 수 있도록 알려준다. 

* 하이퍼 파라미터인 LEARNING_RATE를 처음에 0.01단위로 주다가 손실함수가 더디게 떨어져서 0.1부터 차례로 올려가며 적정범위를 찾아갔다.
