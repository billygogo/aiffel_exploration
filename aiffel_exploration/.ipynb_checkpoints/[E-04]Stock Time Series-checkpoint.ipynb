AR, MA, ARIMA 모두 시계열데이타가 Stationary(안정적인지)의 여부가  중요하므로    
   
 - Stationary :    
  - 과거의 일정한 패턴, 미래에 동일한 반복되는 데이타   
  - 시간에 관계 없이 평균과 분산이 일정한 시계열 데이타   
 - Non-Stationary   
여부를 판단해야 한다.       

* Non-Stationary Series의 경우 Differencing을 해서 Sationary됨을 확인하고 AR, MA, ARIMA인지 판별해   
p,d,q 파라미터를 선별해서 ARIMA를 적용해야 한다.

#### 공통 함수 정의

```python
def plot_rolling_statistics(timeseries, window=12):
    
    rolmean = timeseries.rolling(window=window).mean()  # 이동평균 시계열
    rolstd = timeseries.rolling(window=window).std()    # 이동표준편차 시계열

     # 원본시계열, 이동평균, 이동표준편차를 plot으로 시각화해 본다.
    plt.figure(figsize=(6,4))
    orig = plt.plot(timeseries, color='blue',label='Original')    
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.show(block=False)
    plt.figure(figsize=(6,4))
    std = plt.plot(rolstd, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
from statsmodels.tsa.stattools import adfuller
def augmented_dickey_fuller_test(timeseries):
    # statsmodels 패키지에서 제공하는 adfuller 메서드를 호출합니다.
    dftest = adfuller(timeseries, autolag='AIC')  
    
    # adfuller 메서드가 리턴한 결과를 정리하여 출력합니다.
    print('Results of Dickey-Fuller Test:')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)   
    
```

### (3-1) 정성적 그래프 분석

```python
# 정성적 그래프 분석
plot_rolling_statistics(ts_log, window=12)
```

✒️ 정성적 분석을 통해 해당 데이타가 Stationary 여부를 파악했다. 이동평균 그래프이 일정한 기울기로 상승하지도 않고, 이동표준편차그래프도
일정산 폭을 유지 하지도 못해서 Non-Stationary하다고 판단 된다.   

### (3-2)정량적 분석

```python
#정량적 Augmented Dicky-Fuller Test
augmented_dickey_fuller_test(ts_log)
```

✒️ ADF 테스트에서 귀무가설은 주어진 데이타가 Non-Stationay하다라는 가정이다. ADF 테스트의 결과 p-value가 0.5라는 높은 수치가 나와서    
귀무가설이 기각되지 않고 채택되었다. 따라서 해당 데이타는 Non-Stationary 즉 안정적이지 않은 데이타 이다.    
그러므로 ARIMA 분석을 시도하기 위해서는 Non-Sationary한 데이타를 Stationary 데이타로 변형해야 한다.    
이때 시계열 분해(Time series decomposition)와 차분(Differencing)을 사용해서 데이타를 상쇄 시켜야 한다. 

### (3-3) 시계열 분해

```python
#시계열 분해 (Time Series Decomposition)
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts_log, model='multiplicative', period = 30) 

trend = decomposition.trend # 추세(시간 추이에 따라 나타나는 평균값 변화 )
seasonal = decomposition.seasonal # 계절성(패턴이 파악되지 않은 주기적 변화)
residual = decomposition.resid # 원본(로그변환한) - 추세 - 계절성

plt.figure(figsize=(6,4))
plt.rcParams["figure.figsize"] = (11,6)
plt.subplot(411)
plt.plot(ts_log, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

✒️ Time Series Decomposition을 통해 
* trend:추세-시간 추이에 따라 나타나는 평균값 변화 
* seasonal: 계절성(패턴이 파악되지 않은 주기적 변화)
* __residual__ :원본(로그변환한) - 추세 - 계절성

산출했다.

```python
# Residual 안정성 확인
residual.dropna(inplace=True)
augmented_dickey_fuller_test(residual)
```

✒️ 산출한 Residual 값이 ADF test결과 6.234854e-08로 아주 미세한 값이되었음을 확인했다.    
이것의 의미는 ADF테스트의 귀무가설 즉, 해당데이타는 Non-Stationary하다는 가설이 틀렸다는 것으로 귀무가설은 기각될 수 있고,    
대립가설 즉, 해당 데이타는 Stationary하다는 것을 채택할 수 있는 근거가 될 수 있다.   
여기서는 단지 대립가설이 채택될 근거가 있다는 것 뿐이지 이것 자체가 해당 데이타가 Stationary함을 보장하지는 않는다.

## STEP 5 : 적정 ARIMA 모수 찾기
* ACF, PACF 그려보기 안정성 점검
* 차분 안정성 확인 -> d 구하기
* ACF, PACF 그려보기-> p,q 구하기

### (5-1)ACF, PACF 그려보기

```python
# ACF, PACF 그려보기 -> p,q 구하기
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure(figsize=(6,4))
plot_acf(ts_log)   # ACF : Autocorrelation 그래프 그리기
plot_pacf(ts_log)  # PACF : Partial Autocorrelation 그래프 그리기
plt.show()
```

✒️ ACF그래프가 완만하게 하향되고, PACF는 1이후 급격하게 0으로 수렴함을 보고 이것은 전형적인 Non-Sationary데이타임을 의미한다. 하지만 다음 단계에서 Non-Sationary데이타를 Stationary 데이타를 바꿔주기 위해서 차분(Differenceing)을 사용할 예정이다.

여기 단계에 ACF와 PACF를 형태를 보고, AR, MA, ARIMA인지 판단해서 p와 q를 구해야 나중에 ARIMA에서 order값으로 지정할 수 있다. 
<img src="https://user-images.githubusercontent.com/95423215/227944182-a67a84ac-5c0a-4691-9454-ed82dcdc354a.png">

ACF는 완만하게 하강하면 15이후에 0을 수렴하고 있다. . PACF는 1이후로 급격하세 하강하는 것으로 보아 이 모델은 AR 모델에 가깝다고 이론적으로는 판단된다. p=1로 추정될 수 있다. 그러나 ARIMA모델에 order 파라미터를 찾기위해서는 여러 수치를 대입해보며 모델의 AIC나 MAPE
평가 지표를 보며 적정한 p와 q를 찾아야 한다. 

__일단 이 그래프를 통해서 후보가 될 수 있는 p,q는 (1,0), (0,1), (15, 0) (0,15) 이다.__


Non-Sationary데이타를 Stationary 데이타를 바꿔주기 위해서 차분(Differenceing)을 사용해보자.

### (5-2)차분 안정성 확인 -> d 구하기

```python
# 차분 안정성 확인 -> d 구하기
# 1차 차분 구하기
diff_1 = ts_log.diff(periods=1).iloc[1:]
plt.figure(figsize=(6,4))
diff_1.plot(title='Difference 1st')

augmented_dickey_fuller_test(diff_1)
```

✒️ 1차 차분을 통해서 해당 데이타는 일정한 평균과 분산을 유지하는 Stationary한 데이타가 되었다.  
1차에서 Stationary가 되었으므로 굳이 2차 차분을 시도하지 않았다. 일반적으로 1차 차분으로 충분하면, 2차까지 가능경우도 드물고,    
3차를 시도해야 하는 시계열 데이타라면 그 데이타는 ARIMA모델에 적합하지 않은 모델로 판단하고 다른 모델을 생각해 보아야 한다고 한다.[1]

__1차 차분을 통해서 d = 1임을 알았다.__


## STEP 6 : ARIMA 모델 훈련과 테스트
* ARIMA 모델 빌드와 훈련

## (6-1) ARIMA 모델 빌드와 훈련

```python
from statsmodels.tsa.arima.model import ARIMA

# Build and Train  Model
model = ARIMA(train_data, order=(5, 1,15 )) # 모수는 이전 그래프를 참고 
fitted_m = model.fit() 

print(fitted_m.summary())
```

   
✒️ STEP 5를 통해 구한 p,d,q를 가지고 처음에 (1,1,0), (0,1,1)(0,1,15),(15,1,0) 을 시도하면서  AIC지표를 MAPE지표를 보면서 평가지표 수치를 줄이려 여러 파라미터를 시도했다. 
가장 최적으로 찾는 모델은 __ARIMA(train_data, order=(5, 1,15 ))__ 이다. AIC=-1134.037 , MAPE =1.47%
   

```python
# Forecast : 결과가 fc에 담깁니다. 

fc= fitted_m.forecast(len(test_data), alpha=0.05)  # 95% conf
# Make as pandas series

fc_series = pd.Series(fc, index=test_data.index)   # 예측결과
# Plot

plt.figure(figsize=(9,5), dpi=100)
plt.plot(train_data, label='training')
plt.plot(test_data, c='b', label='actual price')
plt.plot(fc_series, c='r',label='predicted price')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

mse = mean_squared_error(np.exp(test_data), np.exp(fc))
print('MSE: ', mse)

mae = mean_absolute_error(np.exp(test_data), np.exp(fc))
print('MAE: ', mae)

rmse = math.sqrt(mean_squared_error(np.exp(test_data), np.exp(fc)))
print('RMSE: ', rmse)

mape = np.mean(np.abs(np.exp(fc) - np.exp(test_data))/np.abs(np.exp(test_data)))
print('MAPE: {:.2f}%'.format(mape*100))
```

### 시계열 모델의 성능을 평가 지표 ✒️
* AIC(Akaike Information Criterion) :  모형이 데이터를 얼마나 잘 적합시키는지를 나타내는 통계적 측도,   모형 복잡성과 적합도 사이의 균형을 제공하며 값이 낮으면 더 잘 적합됨   
* MAPE(Mean Absolute Percent Error) :시계열의 실제 값과 예측 값 사이의 백분율 차이를 측정하는 오차 메트릭. 모형 예측의 정확도에 대한 측도를 제공하며, 값이 낮을수록 정확도가 높다는 것을 나타냄

```python
ts, np.exp(fc)
```

## STEP 4 : 학습, 테스트 데이터셋 생성

```python
train_data, test_data = ts_log[:int(len(ts_log)*0.9)], ts_log[int(len(ts_log)*0.9):]
plt.figure(figsize=(6,4))
plt.grid(True)
plt.plot(ts_log, c='r', label='training dataset')  # train_data를 적용하면 그래프가 끊어져 보이므로 자연스러운 연출을 위해 ts_log를 선택
plt.plot(test_data, c='b', label='test dataset')
plt.legend()
```

## STEP 5 : 적정 ARIMA 모수 찾기

```python

# ACF, PACF 그려보기 -> p,q 구하기
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure(figsize=(6,4))
plot_acf(ts_log)   # ACF : Autocorrelation 그래프 그리기
plot_pacf(ts_log)  # PACF : Partial Autocorrelation 그래프 그리기
plt.show()
```

✒️ ACF는 24lag 부터 0으로 수렴하면서 천천히 하강하고 있다. PACF는 1이후 부터 급락해 0으로 수렴했다. 
이는 전형적으로 AR의 모델이라 판단되면, p와 q값을 (1,0) (0,1) (0,24),(24,0)로 추정하여 시도해보자. 


```python
# 차분 안정성 확인 -> d 구하기
# 1차 차분 구하기

diff_1 = ts_log.diff(periods=1).iloc[1:]
diff_1.plot(title='Difference 1st')

augmented_dickey_fuller_test(diff_1)

# 2차 차분 구하기
diff_2 = diff_1.diff(periods=1).iloc[1:]
diff_2.plot(title='Difference 2nd')

augmented_dickey_fuller_test(diff_2)
```

✒️ 차분(differncing) 후 평균도 분산도 일정한 흐름을 보여주므로 Sationary 데이타로 판단되며, ADF Test p-value값도 안정됨을 확인했다. 

## STEP 6 : ARIMA 모델 훈련과 테스트

```python
from statsmodels.tsa.arima.model import ARIMA

# Build and Train  Model

model = ARIMA(train_data, order=(0,1,13))#모수는 이전 그래프를 참고 
fitted_m = model.fit() 


print(fitted_m.summary())
```

✒️  ARIMA order=(0,1,13)에서 가장 좋은 평가지표를 보였다. 
AIC = -1148.777, MAPE: 1.47%  

```python
# Forecast : 결과가 fc에 담깁니다. 

fc= fitted_m.forecast(len(test_data), alpha=0.05)  # 95% conf
# Make as pandas series

fc_series = pd.Series(fc, index=test_data.index)   # 예측결과
# Plot

plt.figure(figsize=(6,4), dpi=100)
plt.plot(train_data, label='training')
plt.plot(test_data, c='b', label='actual price')
plt.plot(fc_series, c='r',label='predicted price')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

mse = mean_squared_error(np.exp(test_data), np.exp(fc))
print('MSE: ', mse)

mae = mean_absolute_error(np.exp(test_data), np.exp(fc))
print('MAE: ', mae)

rmse = math.sqrt(mean_squared_error(np.exp(test_data), np.exp(fc)))
print('RMSE: ', rmse)

mape = np.mean(np.abs(np.exp(fc) - np.exp(test_data))/np.abs(np.exp(test_data)))
print('MAPE: {:.2f}%'.format(mape*100))
```

✒️ 예측치에 대한 그래프가 모양은 일직선으로 나왔는데, ARIMA 모수를 조정해서 그래프가 겹쳐지는 위치로 p,q값을 조정해 주었다.
