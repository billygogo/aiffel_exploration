```python
import numpy as np
theta = np.linspace(0,2*np.pi,20)
plt.plot(theta,np.cos(theta))
```

```python
theta = np.linspace(0,2*np.pi,20)
plt.plot(theta,np.cos(theta),'r*--', label='cos')
#plt.figure() #    
plt.plot(theta,np.sin(theta),'go-', label='sin')
plt.xlabel('theta')
plt.ylabel('value')
plt.legend()
plt.title('sin cos')
plt.grid()
#plt.xlim([0,np.pi*2])
#plt.ylim([-1.5,1.5])
plt.axis([0,np.pi*2, -1.5, 1.5])
```

### 2  ,  , 3D 

```python
a = np.linspace(-1,1,100)
b = np.linspace(-1,1,100)

A,B = np.meshgrid(a,b)
L=A**2 + B**2
plt.contour(a,b,L,20)
```

```python
plt.figure(figsize=[10,9])
ax = plt.axes(projection = "3d")
ax.view_init(elev=10, azim=-45) # , 
ax.plot_surface(A,B,L)


plt.figure(figsize=[10,9])
ax = plt.axes(projection = "3d")
ax.plot_wireframe(A,B,L)
```

```python
from matplotlib.colors import LightSource
plt.figure(figsize=[10,9])
ax = plt.axes(projection = "3d")
#ax.view_init(elev=10, azim=-45)
ls = LightSource(altdeg = 20, azdeg=-45)
rgb = ls.shade(L,plt.cm.RdYlBu)
ax.plot_surface(A,B,L, facecolors=rgb)
```

```python
from matplotlib.colors import LightSource
plt.figure(figsize=[10,9])
ax = plt.axes(projection = "3d")

ax.plot_surface(A,B,L, cmap='viridis')
# viridis , plasma, inferno, magma , cividis
```

# scatterplot

```python
a = np.linspace(-1,1,20)
b = np.linspace(-1,1,20)

A,B = np.meshgrid(a,b)
L=A**2 + B**2

plt.figure(figsize=[10,9])
ax = plt.axes(projection = "3d")
ax.plot(A.reshape(-1,1).squeeze(),B.reshape(-1,1).squeeze(),L.reshape(-1,1).squeeze(),"ro" )

```

```python
a = np.linspace(-1,1,20)
b = np.linspace(-1,1,20)

A,B = np.meshgrid(a,b)
L=A**2 + B**2

plt.figure(figsize=[10,9])
ax = plt.axes(projection ="3d")
ax.scatter(A,B,L, s=15, c ="g",marker="*", edgecolor='r')
```

```python
plt.figure(figsize=[10,9])
ax = plt.axes(projection = "3d")

ax.plot_surface(A,B,L, alpha =0.3, cmap='viridis')
ax.scatter(A,B,L, s = 10, c='g', marker="^")
```

# plotly

```python
import plotly.graph_objects as go
```

```python
a = np.linspace(-2,2,100)
b = np.linspace(-2,2,100)
A,B = np.meshgrid(a,b)
L = np.cos(A**2)+ np.sin(B**2.4)

fig = go.Figure(data=[go.Surface(x=a, y=b,z=L, colorscale="viridis", opacity=0.5)])
fig.update_traces(contours_z=dict(show=True,usecolormap=True, highlightcolor="limegreen",project_z=True))
fig.update_layout(title="cos(A**2)+ sin(B**2.4", width=700, height=600)
```

```python
fig = go.Figure(data=[go.Surface(x=a, y=b,z=L, opacity=0.5)])
fig.update_traces(contours_z=dict(show=True,usecolormap=True, highlightcolor="limegreen",project_z=True))
fig.update_layout(title="cos(A**2)+ sin(B**2.4", width=700, height=600, scene = dict(zaxis=dict(nticks=40, range=[-3,3])))
```
