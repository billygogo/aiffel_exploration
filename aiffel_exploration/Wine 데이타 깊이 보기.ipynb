## 3. 시각화시켜 features의 관련성 탐색

```python
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format = 'retina' 
```

```python
plt.scatter(wine_frame.data['alcohol'], wine_frame.data['proline'])
plt.show()
```

alcohol과 proline는 관계를 scatter 그래프로 관찰해 보았다. 

```python
plt.scatter(wine_frame.data['total_phenols'], wine_frame.data['proline'])
plt.show()
```

```python
plt.hist(wine_frame.data['magnesium'])
```

```python
plt.hist(wine_frame.data['proline'])
```

```python
# 축 그리기
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

# alcalinity_of_ash , color_intensity
# 그래프 그리기
patches = ax1.hist(wine_frame.data['alcalinity_of_ash'], bins=50, density=False) #bins는 x값을 총 50개 구간으로 나눈다는 뜻입니다.
patches = ax1.hist(wine_frame.data['proline'], bins=50, density=False, alpha=0.5)
ax1.xaxis.set_ticks_position('bottom') # x축의 눈금을 아래 표시 
ax1.yaxis.set_ticks_position('left') #y축의 눈금을 왼쪽에 표시

# 라벨, 타이틀 달기
plt.xlabel('alcalinity_of_ash')
plt.ylabel('proline')
ax1.set_title('alcalinity_of_ash & proline')

# 보여주기
plt.show()
```

proline 데이타가 다른 데이타들과 scale 달라서 표준편차가 크고,  Normalization이 필요한지 검토 필요

# Wine 데이타 개요

총 6개의 정보 :'data', 'target', 'frame', 'target_names', 'DESCR', 'feature_names'
* data: 178 rows × 13  columns 
* target: 분류해야 할 class 'class_0', 'class_1', 'class_2'
* feature_names : data의 column 이름


### 회고

* Wine 데이타의 경우 다른 Feature과 비교하며 데이타를 시각화 시켜 관찰해 보았다. 
* 관찰 결과 proline 데이타가 다른 데이타들 보다 심하게 scale이 차이가 나고 있다고 보여진다.
